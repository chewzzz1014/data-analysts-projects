## Import Data
The data were downloaded on my local folder. Let's import them 

```python
import os
abs_path = os.path.abspath("./divvy-tripdata")
p = os.listdir("./divvy-tripdata")
p
```

```python
files_path = [ abs_path + "\\" +f for f in p[:len(p)-1] ]
files_path
```

```python
# put all dfs in a list
df_202109 = pd.read_csv(files_path[0])
df_202110 = pd.read_csv(files_path[1])
df_202111 = pd.read_csv(files_path[2])
df_202112 = pd.read_csv(files_path[3])

df_202201 = pd.read_csv(files_path[4])
df_202202 = pd.read_csv(files_path[5])
df_202203 = pd.read_csv(files_path[6])
df_202204 = pd.read_csv(files_path[7])
df_202205 = pd.read_csv(files_path[8])
df_202206 = pd.read_csv(files_path[9])
df_202207 = pd.read_csv(files_path[10])
df_202208 = pd.read_csv(files_path[11])

dfs = [df_202109,
        df_202110,
        df_202111,
        df_202112,
        df_202201,
        df_202202,
        df_202203,
        df_202204,
        df_202205,
        df_202206,
        df_202207,
        df_202208
        ]
```

## Process

```python
df_202201.head() 
```

```python
# convert to date time
for df in dfs:
    df["started_at"] = pd.to_datetime(df["started_at"])
    df["ended_at"] = pd.to_datetime(df["ended_at"])
```

```python
## create a column called “ride_length” by subtracting the column “started_at” from the column “ended_at” 
for df in dfs:
    df["ride_length"] = df["ended_at"] - df["started_at"]
df_202201.head(2)
```

```python
# add column “day_of_week” represent the day of the week that each ride 
# note what we want:  1 = Sunday and 7 = Saturday

'''
    from to
sun 6 1
mon 0 2
tue 1 3
wed 2 4
thu 3 5
fri 4 6
sat 5 7
'''

# pandas return monday = 0 and sunday = 6. So, add each weekday by one
for df in dfs:
    weekday = df["started_at"].dt.weekday.apply(lambda x: 1 if x==6 else x+2 )
    df["day_of_week"] = weekday

df_202201.head()
```

```python
# combine start_lat and start_lng into start_coordinates
# same for end_at and end_lng

for df in dfs:
    df["start_coordinate"] = df['start_lat'].astype(str) + "," + df['start_lng'].astype(str)
    df.drop(["start_lat", "start_lng"], inplace=True, axis=1)
    
    df["end_coordinate"] = df['end_lat'].astype(str) + "," + df['end_lng'].astype(str)
    df.drop(["end_lat", "end_lng"], inplace=True, axis=1)

df_202201.head()
```

```python
# convert member to 1 and casual to 0
for df in dfs:
    df["member_casual"] = df["member_casual"].apply(lambda x: 1 if x=="member" else 0)

df_202201.head(3)
```

## Mean of ride_length

```python
print("Mean of ride length, from Sep 2021 to Aug 2022\n")
for df in dfs:
    mean_ride_length = df.ride_length.mean()
    print(f"{df.started_at.dt.month_name()[0]} {df.started_at.dt.year[0]} : {mean_ride_length}")
```

## Max ride_length

```python
print("Max of ride length, from Sep 2021 to Aug 2022\n")
for df in dfs:
    max_ride_length = df.ride_length.max()
    print(f"{df.started_at.dt.month_name()[0]} {df.started_at.dt.year[0]} : {max_ride_length}")
```

## Mode of day_of_week

```python
print("Mode of day of week, from Sep 2021 to Aug 2022\n")

for df in dfs:
    mode_day_of_week = df.day_of_week.mode()
    print(f"{df.started_at.dt.month_name()[0]} {df.started_at.dt.year[0]} : {mode_day_of_week}")
```
